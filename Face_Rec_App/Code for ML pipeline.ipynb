```python
# process all img:
def extract_images(path,gender,i):
    img=cv2.imread(path) # read the path
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # convert to gray
    faces=haar.detectMultiScale(gray,1.5,5) # identify img
    for x,y,w,h in faces:
        roi=img[y:y+h,x:x+w] # crop
        if gender=='male':
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
# process img: 
def resize_img(path_to_resize): 
    try:
        # 1. read img
        img=cv2.imread(path_to_resize)
        # 2. convery to gray
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        # 3. resize to 100*100
        size=gray.shape[0]
        if size>=100:
            gray_re=cv2.resize(gray,(100,100),cv2.INTER_AREA)  # shrink
        else:
            gray_re=cv2.resize(gray,(100,100),cv2.INTER_CUBIC) # enlarge
        # 4. flatten img 1*10,000
        return gray_re.flatten()
    except:
        return None
```

# PCA

```python
X1=X-X.mean(axis=0)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=None,whiten=True,svd_solver='auto')
```

```python
x_pca=pca.fit_transform(X1)
```

```python
x_pca.shape  # 'auto' reduces the component to the row size
```

```python
eigen_ratio=pca.explained_variance_ratio_
eigen_ratio_cum=np.cumsum(eigen_ratio)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(eigen_ratio[:200],'r>--') # visualize first 200 components
plt.xlabel('number of components')
plt.ylabel('Explained Variance Ratio') # 1st component explains 25%, sec 12%, third 10% ,etc
plt.subplot(1,2,2)
plt.plot(eigen_ratio_cum[:200],'r>--')
plt.xlabel('number of components')
plt.ylabel('Cumulative Explained Variance Ratio')  # 50 components explain 80% variance
plt.show()
```

## Conclusion: using elbow methid, consider num of components is between 25-30
## Since if considering components between 25-30, the explained var is around 75%, 
## In order to get min 80% var, considering 50 components instead

```python
pca_50=PCA(n_components=50,whiten=True,svd_solver='auto')
x_pca_50=pca_50.fit_transform(X1)
```

```python
x_pca_50.shape
```

# Make Pipeline Model

```python
import pickle
```

```python
# pickle files:
mean=pickle.load(open('./model/mean_preprocess.pickle','rb'))
model_svm=pickle.load(open('./model_svm.pickle','rb'))
model_pca=pickle.load(open('./model/pca_50.pickle','rb'))
print('Model Load Successfully')
```

```python
gender_pre=['Male','Female']
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python


def pipeline_model(img,color='rgb'):
    # 2. convery to gray
    if color=='bgr':
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    else:
        gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    # 3. crop using haar
    faces=haar.detectMultiScale(gray,1.5,3)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2) # draw rectangle
        roi=gray[y:y+h,x:x+w] # crop
    # 4. normalization
        roi=roi/255.0
    # 5. resize into (100,100)
        if roi.shape[1]>100:
            roi_resize=cv2.resize(roi,(100,100),cv2.INTER_AREA)
        else:
            roi_resize=cv2.resize(roi,(100,100),cv2.INTER_CUBIC)
    # 6. flatten: (1*10000)
        roi_reshape=roi_resize.reshape(1,10000)    # row wise
    # 7. subtract the mean
        roi_mean=roi_reshape-mean
    # 8. get eigen image
        eigen_image=model_pca.transform(roi_mean)
    # 9. pass to ML model (svm)
        results=model_svm.predict_proba(eigen_image)[0]
    # 10. 
        predict=results.argmax() # 0 or 1
        score=results[predict]
    # 11. 
        text='%s : %0.2f'%(gender_pre[predict],score)
        cv2.putText(img,text,(x,y),font,1,(0,255,0),2)
    return img
```

```python
# test data
test_data_path='./data/male_000281.jpg'
color='bgr'
# 1. read img
img=Image.open(test_data_path)
# 2. convert to array
img=np.array(img)
# 3. pass to pipeline model
img=pipeline_model(img)
plt.imshow(img)
```
